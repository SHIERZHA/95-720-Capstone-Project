## Google API Call ##
You will need to create an account and apply for a key at  
https://developers.google.com/maps/documentation/distance-matrix/start

```python
def calDistance(origin, destination):

    payload = {"units": "imperial",
    "origins": origin,
    "destinations": destination,
    "key": "", # You will need to create an account and apply for a key at  
               # https://developers.google.com/maps/documentation/distance-matrix/start
    "language": "en-US"}

    api_url = 'https://maps.googleapis.com/maps/api/distancematrix/json'

    result = requests.get(api_url, params=payload)

    res = result.json()
    
#     print(res)

    try:
        dist = round(float(res['rows'][0]['elements'][0]['distance']['value']) * 0.000621371,2)
    except:
        print(res)
        dist = -1

    return dist
```

```python
def cal_distance_for_county(name, butler):
    distance = []
    for i in range(len(butler)):
        origin =  butler.loc[i,"origin"] + ", " + butler.loc[i,"origin_zip"]
        destination = butler.loc[i, "destination"] + ", " + butler.loc[i,"destination_zip"]
        distance.append(calDistance(origin, destination))
#         print("origin ", origin, "destination", destination)
    butler["DISTANCE"] = distance
    print(name, " Done.")
```
